### Scrape Data

```python
with open("Asterank.html", "r") as rank:
    soup=BeautifulSoup(rank)
```

```python
# header

header = [th.getText().strip() for th in table.find_all('thead')[0].find_all('th')]
header
```

```python
# separate each row in own list

data_rows = [[td.get_text().strip() for td in tr.find_all('td')] for tr in all_rows]

data_rows
```

```python
rank = pd.DataFrame(data_rows, columns = header)
rank.head(10)
```

### 2. Regular expression to remove bad data

The table on the website has internal calculations to display (PHA) in column orbit_class (Group) as True or not. When scrapping the website, the data pulls (PHA) for all instances even if it should not have it. Using regular expressions, the (PHA) will be removed from the orbit_class column.

```python
rank.orbit_class = (rank.orbit_class.replace(r'[(PHA)]+$', '', regex=True))
```

```python
def strip_whitespace(col): 
    return col.strip()

rank['orbit_class'] = rank['orbit_class'].apply(strip_whitespace)
```

### 4. Change dollars columns from strings to integer dollar figures


Remove the words 'million," "billion," and "trillion" and replace with numeric figures.

#### value_dollars

```python
split = rank['value_dollars'].str.split(" ", n=1, expand = True)
```

```python
header = {0:'number', 1:'value'}
```

```python
split.rename(columns = header, inplace=True)
```

```python
split.head()
```

```python
for type in [split.value]:
    print(type.unique())
```

```python
# create duplicate column for reference

rank['value_dollars2']=rank['value_dollars']
```

```python
# dictionary for columns over 100 trillion, change to 101 trillion

over = {'>100 trillion': '101 trillion'}
```

```python
# where over is matched in value_dollars, replace with over

rank.replace({'value_dollars':over}, inplace=True)
```

```python
rank.iloc[500:505]
```

```python
# dictionary for what each value represents

repl_dict = {'million':'*1e6', 'billion':'*1e9', 'trillion':'*1e12'}
```

```python
# where the dictonary is found, replace and evaluate to integer value

rank['value_dollars'] = rank['value_dollars'].replace(repl_dict, regex=True).map(pd.eval)
```

```python
rank
```

```python
rank.iloc[980:1000]
```

#### est_profit_dollars

```python
# create copy for refrence

rank['est_profit_dollars2']=rank['est_profit_dollars']
```

```python
# replace >100 trillion with over dict

rank.replace({'est_profit_dollars':over}, inplace=True)
```

```python
# replace all string values with integers

rank['est_profit_dollars'] = rank['est_profit_dollars'].replace(repl_dict, regex=True).map(pd.eval)
```

```python
rank
```

```python
# remove duplicated columns, no longer needed

rank.drop(labels = ['value_dollars2', 'est_profit_dollars2'], axis=1, inplace=True)
```

### 5. Binning

Place the value of each asteroid and estimated profit into bins of no value, thousands, millions, billions, trillions, and over trillions.

```python
bins = [0, 1000, 1000000, 1000000000, 1000000000000, 100000000000000, 102000000000000]
```

```python
bin_names = ['no value', 'thousands', 'millions', 'billions', 'trillions', 'over trillions']
```

#### value_dollars bin

```python
value_bin = pd.Series(rank.value_dollars, name = 'value')
```

```python
value = pd.cut(value_bin, bins, labels=bin_names, right=False)
```

```python
pd.value_counts(value)
```

```python
# add the bins into the dataframe

rank = pd.concat([rank, value], axis=1)
```

```python
rank
```

#### est_profit_dollars bin

```python
profit_bin = pd.Series(rank.est_profit_dollars, name = 'profit')
```

```python
profit = pd.cut(profit_bin, bins, labels=bin_names, right=False)
```

```python
pd.value_counts(profit)
```

```python
rank = pd.concat([rank, profit], axis = 1)
```

```python
rank.head()
```

```python
# move position of value and profit columns in dataframe

value = rank['value']
rank.drop(labels=['value'], axis=1, inplace=True)
rank.insert(5, 'value', value)

profit = rank['profit']
rank.drop(labels=['profit'], axis=1, inplace=True)
rank.insert(7, 'profit', profit) 
```
